```python
import pandas as pd
df_disney = pd.read_csv("../disney_dataset (2).csv")
df_disney.head()
```

```python
correlation = df_disney['imdb'].corr(df_disney['box_office'])
print(f"Correlation between IMDb ratings and box office: {correlation}")
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(df_disney['imdb'], df_disney['box_office'])
plt.xlabel('IMDb Ratings')
plt.ylabel('Box Office Revenue')
plt.title('Scatter Plot of IMDb Ratings vs. Box Office Revenue')
plt.show()
```

```python
import numpy as np

plt.scatter(df_disney['imdb'], df_disney['box_office'], label='Data Points')

# Calculate the regression line
slope, intercept = np.polyfit(df_disney['imdb'], df_disney['box_office'], 1)
regression_line = slope * df_disney['imdb'] + intercept

# Plot the regression line
plt.plot(df_disney['imdb'], regression_line, color='red', label='Regression Line')

# Add labels and title
plt.xlabel('IMDb Ratings')
plt.ylabel('Box Office Revenue (Millions)')
plt.title('Scatter Plot of IMDb Ratings vs. Box Office Revenue')
plt.legend()

# Show plot
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

# Scatter plot
plt.scatter(df_disney['imdb'], df_disney['box_office'], label='Data Points')

# Calculate the regression line
slope, intercept = np.polyfit(df_disney['imdb'], df_disney['box_office'], 1)
regression_line = slope * df_disney['imdb'] + intercept

# Plot the regression line
plt.plot(df_disney['imdb'], regression_line, color='red', label='Regression Line')

# Formatting function for y-axis labels to show millions
def millions(x, pos):
    return '%1.1fM' % (x * 1e-6)

# Apply the formatter to the y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))

# Add labels and title
plt.xlabel('IMDb Ratings')
plt.ylabel('Box Office Revenue (Millions)')
plt.title('Scatter Plot of IMDb Ratings vs. Box Office Revenue')
plt.legend()

# Show plot
plt.show()
```
