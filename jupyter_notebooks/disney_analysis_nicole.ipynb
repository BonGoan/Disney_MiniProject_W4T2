```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df_disney = pd.read_csv("..\data\disney_dataset_analysis.csv")
df_disney
```

```python
correlation = df_disney ['imdb'].corr(df_disney['running_time'])
print (correlation)
```

```python
sns.lmplot(x='running_time', y='imdb', data=df_disney, aspect=2, height=6)
plt.title('Correlation between IMDb Ratings and Running Time')
plt.xlabel('Running Time (minutes)')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
#relationship is very weak between running time and ratings
#slight upward slope of the regression line - a slight tendency for longer movies to receive higher IMDb ratings
#wide spread of points - other factors are likely more important in determining IMDb ratings than just the length of the movie.
```

```python
sns.lmplot(x='running_time', y='imdb', data=df_disney, aspect=2, height=6, line_kws={'color': 'red'})

# Add title and labels
plt.title('Correlation between IMDb Ratings and Running Time')
plt.xlabel('Running Time (minutes)')
plt.ylabel('IMDb Rating')

# Save the image
plt.savefig('imdb_vs_running_time.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
# Box plot for IMDb ratings
plt.figure(figsize=(8, 6))
sns.boxplot(y='imdb', data=df_disney)
plt.title('Box Plot of IMDb Ratings')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
# Box plot for Running Time
plt.figure(figsize=(8, 6))
sns.boxplot(y='running_time', data=df_disney)
plt.title('Box Plot of Running Time')
plt.ylabel('Running Time (minutes)')
plt.show()
```

```python
def categorize_rating(rating):
    if rating < 5:
        return 'poor rating'
    elif 5 <= rating <= 7:
        return 'moderate rating'
    else:
        return 'good rating'
df_disney['rating_category'] = df_disney['imdb'].apply(categorize_rating)

(df_disney)
```

```python
def categorize_running_time(runtime):
    if runtime < 120:
        return 'below 120 minutes'
    else:
        return '120 minutes or more'
df_disney['running_time_category'] = df_disney['running_time'].apply(categorize_running_time)

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='running_time_category', y='imdb', data=df_disney)
plt.title('Box Plot of IMDb Ratings by Running Time Category')
plt.xlabel('Running Time Category')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
#The plot suggests that while longer movies generally receive more consistent and slightly higher ratings, 
# shorter movies have a more diverse range of ratings, including some lower-rated films.
# longer movies tend to be more favorably received or possibly that they are more consistent in quality, 
# while shorter movies might vary more widely in audience reception.
```

```python
df_below_120 = df_disney[df_disney['running_time'] < 120]

# Plot the KDE plot for IMDb ratings of movies below 120 minutes
plt.figure(figsize=(10, 6))
sns.kdeplot(df_below_120['imdb'], shade=True, color="blue")
plt.title('KDE Plot of IMDb Ratings for Movies with Running Time Below 120 Minutes')
plt.xlabel('IMDb Rating')
plt.ylabel('Density')
plt.show()
```

```python
df_hypo3 = df_disney.groupby('based_on_book')['imdb'].mean().reset_index()
df_hypo3.columns = ['Based on Book', 'Average IMDb Rating']
display(df_hypo3)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Based on Book', y='Average IMDb Rating', data=df_hypo3, palette='muted')
plt.title('Average IMDb Ratings: Based on Books vs. Not Based on Books')
plt.xlabel('Based on Book')
plt.ylabel('Average IMDb Rating')
plt.ylim(df_hypo3['Average IMDb Rating'].min() - 0.5, df_hypo3['Average IMDb Rating'].max() + 0.5)  # Adjust y-axis limits
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
df_books = df_disney.groupby(['based_on_book', 'rating_category']).size().reset_index(name='count')
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='based_on_book', y='count', hue='rating_category', data=df_books, palette='deep')
plt.title('Comparison of Rating Categories: Based on Books vs. Not Based on Books')
plt.xlabel('Based on Book')
plt.ylabel('Count')
plt.legend(title='IMDb Rating Category')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('Comparison of Ratings with books.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.scatter(df_disney['imdb'], df_disney['box_office'], label='Data Points')
# Calculate the regression line
slope, intercept = np.polyfit(df_disney['imdb'], df_disney['box_office'], 1)
regression_line = slope * df_disney['imdb'] + intercept
# Plot the regression line
plt.plot(df_disney['imdb'], regression_line, color='red', label='Regression Line')
# Add labels and title
plt.xlabel('IMDb Ratings')
plt.ylabel('Box Office Revenue (Millions)')
plt.title('Scatter Plot of IMDb Ratings vs. Box Office Revenue')
plt.legend()
# Show plot
plt.show()
```
