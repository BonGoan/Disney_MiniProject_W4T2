```python
df_books = pd.DataFrame(df['based_on'])

# Extract book name and author
def extract_book_author(entry):
    if pd.isna(entry):
        return np.nan, np.nan

    try:
        # Try to parse as list
        parsed = ast.literal_eval(entry)
        if isinstance(parsed, list):
            # Flatten the list into a single string and then split by ' by '
            combined_entry = ' '.join(parsed)
            parts = combined_entry.split(' by ', 1)  # Split by ' by ', only at the first occurrence
            if len(parts) == 2:
                book_name = parts[0].replace('"', '').strip()
                author = parts[1].replace('"', '').strip()
                return book_name, author
            else:
                return combined_entry, np.nan
        else:
            return np.nan, np.nan
    except:
        # If entry is not in list format
        if ' by ' in entry:
            parts = entry.split(' by ', 1)  # Split by ' by ', only at the first occurrence
            book_name = parts[0].replace('"', '').strip()
            author = parts[1].replace('"', '').strip() if len(parts) > 1 else np.nan
            return book_name, author
        else:
            return entry, np.nan

# Apply the function to create new columns
df_books['book_name'], df_books['author'] = zip(*df_books['based_on'].apply(extract_book_author))

display(df_books[['based_on', 'book_name', 'author']])
```

```python
# Define regex pattern for special characters
special_characters_pattern = re.compile(r'[!@#$%^&*()_+{}|:"<>?~]')

# Check for special characters in a DataFrame
def check_dataframe_for_special_chars(df):
    for column in df.columns:
        df[column] = df[column].astype(str)  # Convert all columns to string for searching
        special_chars_found = df[column].apply(lambda x: bool(special_characters_pattern.search(x)))
        if special_chars_found.any():
            print(f"Special characters found in column '{column}'")
            print(df[special_chars_found][column])

# Check the DataFrame for special characters
check_dataframe_for_special_chars(df)
```

```python
# Replace special characters with an empty space
def replace_special_chars(text):
    return special_characters_pattern.sub(' ', text)

# Iterate over each column and apply the replacement function
for column in df.columns:
    df[column] = df[column].map(lambda x: replace_special_chars(str(x)) if isinstance(x, str) else x)
```
